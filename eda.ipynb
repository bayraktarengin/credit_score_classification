# Creating the model

```python
df["Credit_Mix"].value_counts()
```

```python
# There are too much unknown variables "NM". We can consider them as missing values. 
df["Payment_of_Min_Amount"].value_counts()
```

```python
df["Credit_Mix_Num"] = df["Credit_Mix"].map({"Standard": 1, "Good": 2, "Bad": 0})
```

```python
feautures = ['Annual_Income', 'Monthly_Inhand_Salary', 'Num_Bank_Accounts', 'Num_Credit_Card', 'Interest_Rate', 'Num_of_Loan', 'Delay_from_due_date', 'Num_of_Delayed_Payment', 'Credit_Mix_Num', 'Num_Credit_Inquiries', 'Outstanding_Debt', 'Credit_History_Age', 'Monthly_Balance']
target = "Credit_Score"
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import metrics  
```

```python
x = df[feautures]
y = df[target]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# creating a RF classifier
clf = RandomForestClassifier(n_estimators = 100)  
  
# Training the model on the training dataset
# fit function is used to train the model using the training sets as parameters
clf.fit(x_train, y_train)
  
# performing predictions on the test dataset
y_pred = clf.predict(x_test)
  
# metrics are used to find accuracy or error
from sklearn import metrics  
print()
  
# using metrics module for accuracy calculation
print("ACCURACY OF THE MODEL: ", metrics.accuracy_score(y_test, y_pred))
```

```python
for n_estimator in [100,400,600]:
    clf = f"clf_{n_estimator}"
    clf = RandomForestClassifier(n_estimators = n_estimator)  
  
    # Training the model on the training dataset
    # fit function is used to train the model using the training sets as parameters
    clf.fit(x_train, y_train)

    # performing predictions on the test dataset
    y_pred = clf.predict(x_test)


    # using metrics module for accuracy calculation
    print(f"ACCURACY OF THE MODEL FOR {n_estimator}: ", metrics.accuracy_score(y_test, y_pred))
```
